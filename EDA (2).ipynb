## Hypothes testing:

```r
# Define the mean life expectancy for each continent
df3 <- gapminder %>% 
  group_by(continent) %>% 
  summarize(meanlife = mean(lifeExp, na.rm = TRUE))

# Create a density plot of life expectancy by continent 
#note we will add a vertical line to indicate the mean life expectancy for each continent
ggplot(data = gapminder, aes(x = lifeExp, fill = continent)) + 
  geom_density(alpha = 0.5) +
  geom_vline(data = df3, aes(xintercept = meanlife), color = "black", linetype = "dashed") +
  labs(title = "Density Plot of Life Expectancy by Continent", 
       x = "Life Expectancy", y = "Density") +
  scale_fill_discrete(name = "Continent") +
  facet_wrap(~ continent) +
  theme_bw(base_size = 20) +
  theme(plot.title = element_text(hjust =0.5 ),
        axis.title = element_text(size = 18),
        strip.text = element_text(size = 20),
        legend.position = "bottom",
        axis.text = element_text(size = 14),
        axis.line = element_line(color = "black"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        panel.background = element_blank()) +
  scale_x_continuous(breaks = seq(0, 100, by = 10))+
  labs(caption = "The vertical line to indicate the mean life expectancy for each continent")
```

```r
# Single sample t-test 
# Hypothes testing: 
#the null Hypothes----> HO: the mean 1ife expectancy in Africa is 50 years
# vs                    Hl: the mean 1ife expectancy in Africa is not 50 years
```

```r
gapminder %>%
filter(continent == "Africa") %>%
select(lifeExp) %>%
t.test(mu = 50)#population mean it's equal to 50 in other words the assumed mean population mean is 50 and it's testing whether or not our sample data finds that is a statistically significant
```

```r
# Perform a hypothesis test to compare life expectancy between continents
# H0= both the countries have the same life expectancy
asia_europe_lifeExp <- gapminder %>%
  filter(continent %in% c("Africa", "Americas")) %>%
  select(continent, lifeExp)

t.test(lifeExp ~ continent, data = asia_europe_lifeExp)
# p value is very small 
#we can reject the notion that the two means of the same and we can accept the fact that the different this difference is statistically significant 
```

```r
# Perform a hypothesis test to compare life expectancy between continents
# H0= both the countries have the same life expectancy
asia_europe_lifeExp <- gapminder %>%
  filter(continent %in% c("Asia", "Europe")) %>%
  select(continent, lifeExp)

t.test(lifeExp ~ continent, data = asia_europe_lifeExp)
# p value is very small 
#we can reject the notion that the two means of the same and we can accept the fact that the different this difference is statistically significant 
```

```r
# Perform a hypothesis test to compare life expectancy between continents
# H0= both the countries have the same life expectancy
asia_europe_lifeExp <- gapminder %>%
  filter(continent %in% c("Oceania", "Europe")) %>%
  select(continent, lifeExp)

t.test(lifeExp ~ continent, data = asia_europe_lifeExp)

# the p value 0.006415
# statistically significant
# we can reject our null hypothesis 
#Typically, a p-value of less than 0.05 is considered statistically significant, which means that there is strong evidence against the null hypothesis and we can reject it. However, in your example, the p-value is 0.006415, which is smaller than 0.05, so we can reject the null hypothesis. This means that there is a statistically significant difference between the mean life expectancy of Europe and Oceania.
```
